```python
# Importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor, plot_tree
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.preprocessing import MinMaxScaler

# Step 1: Create a dataset with house features and prices
data = {
    "Size (sq ft)": [1200, 1500, 1800, 2500, 1100, 1400, 1600, 2000, 2200, 3000,
                     800, 1000, 2700, 1900, 2100, 1700, 2300, 3200, 2800, 3500],
    "Bedrooms": [2, 3, 3, 4, 2, 3, 3, 4, 4, 5,
                 1, 2, 5, 3, 4, 3, 4, 5, 4, 5],
    "Age (years)": [10, 15, 20, 5, 25, 30, 15, 10, 8, 4,
                    50, 45, 3, 12, 7, 20, 6, 2, 9, 1],
    "Price (USD)": [200000, 250000, 300000, 400000, 150000, 220000, 270000, 350000, 370000, 500000,
                    100000, 120000, 480000, 320000, 400000, 290000, 450000, 600000, 500000, 700000]
}

# Convert the data dictionary into a Pandas DataFrame
df = pd.DataFrame(data)

# Step 2: Define features (X) and target variable (y)
X = df[["Size (sq ft)", "Bedrooms", "Age (years)"]]  # Features
y = df["Price (USD)"]  # Target variable

# Step 3: Normalize features using Min-Max Scaling
scaler = MinMaxScaler()
X_normalized = scaler.fit_transform(X)

# Split the normalized data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_normalized, y, test_size=0.2, random_state=42)

# Step 4: Define the hyperparameter grid
param_grid = {
    "max_depth": [2, 4, 6, 8, 10, None],
    "min_samples_split": [2, 5, 10],
    "min_samples_leaf": [1, 2, 4]
}

# Step 5: Perform GridSearchCV to find the best hyperparameters
grid_search = GridSearchCV(
    estimator=DecisionTreeRegressor(random_state=42),
    param_grid=param_grid,
    scoring="neg_mean_squared_error",  # Use MSE as the scoring metric
    cv=5,  # 5-fold cross-validation
    n_jobs=-1  # Use all available cores
)

# Fit GridSearchCV on the training data
grid_search.fit(X_train, y_train)

# Retrieve the best model and parameters
best_model = grid_search.best_estimator_
best_params = grid_search.best_params_

# Step 6: Evaluate the best model on the test set
y_pred = best_model.predict(X_test)

# Calculate metrics
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Display the best hyperparameters and model performance
print("Best Hyperparameters:")
print(best_params)
print("\nModel Performance Metrics:")
print(f"Mean Squared Error (MSE): {mse}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"Mean Absolute Error (MAE): {mae}")
print(f"R-squared (R^2): {r2}")

# Step 7: Predict the price for a new house using the best model
new_house = [[2000, 3, 10]]  # Features of the new house
new_house_normalized = scaler.transform(new_house)  # Normalize the new house features
predicted_price = best_model.predict(new_house_normalized)
print("\nPredicted Price for the new house (2000 sq ft, 3 bedrooms, 10 years old):", predicted_price[0])

# Step 8: Visualize the decision tree
plt.figure(figsize=(16, 10))
plot_tree(
    best_model,
    feature_names=["Size (sq ft)", "Bedrooms", "Age (years)"],
    filled=True,
    rounded=True,
    fontsize=10
)
plt.title("Decision Tree Visualization")
plt.show()
```

 

Raíz del árbol (nodo inicial):

El árbol comienza con una regla de división basada en una característica. En este caso, el nodo raíz decide si el número de "Bedrooms" (normalizado) es menor o igual a 0.625.
Los datos se dividen en dos ramas:
True (izquierda): Si la condición se cumple.
False (derecha): Si la condición no se cumple.
Nodos intermedios:

Cada nodo intermedio contiene:
La regla de división basada en una característica.
squared_error: El error cuadrático medio acumulado para los datos en ese nodo.
samples: La cantidad de datos que llegan a este nodo.
value: El promedio de los valores objetivo (precio en este caso) de los datos en ese nodo.
Hojas del árbol (nodos finales):

Las hojas representan las predicciones finales del modelo.
Cada hoja tiene un valor fijo (value), que es el promedio de los precios de las casas que cayeron en ese nodo durante el entrenamiento.
Por ejemplo, una hoja con value = 120000.0 significa que el modelo predice un precio de 120,000 para cualquier casa cuyas características lleguen a este nodo.
Flujo del árbol:

Al evaluar un dato, el modelo sigue las condiciones de cada nodo, desde la raíz hasta una hoja.
Por ejemplo:
Si una casa tiene 4 dormitorios, el modelo tomará el camino "False" en el nodo raíz porque 4 dormitorios (normalizado) no cumple con la condición Bedrooms <= 0.625.
El modelo continuará evaluando los nodos intermedios hasta llegar a una hoja con un valor predefinido.
Predicción predefinida:

Sí, cada nodo hoja ya tiene un valor predefinido (promedio de los datos en ese nodo).
En la práctica, cuando una nueva casa pasa por el árbol, el modelo simplemente devuelve el valor de la hoja correspondiente como predicción.
