```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Paso 1: Generar un dataset aleatorio con m√∫ltiples columnas num√©ricas
np.random.seed(13)
data = {
    'A': np.random.randint(13, 300, size=100),
    'B': np.random.normal(loc=64, scale=8, size=100),
    'C': np.random.uniform(low=125, high=8, size=100),
    'D': np.random.exponential(scale=2.3, size=100)
}

#    'A': np.random.randint(1, 100, size=100),
#    'B': np.random.normal(loc=50, scale=10, size=100),
#    'C': np.random.uniform(low=10, high=60, size=100),
#    'D': np.random.exponential(scale=2, size=100)


# Crear un DataFrame
df = pd.DataFrame(data)

# Mostrar los primeros 5 registros del dataset original

print("Gabriel Guzm√°n - seed(13) - c-G.G\n")
print("Dataset original:\n", df.head())

# Paso 2: Funci√≥n para aplicar la normalizaci√≥n Z-Score a todas las columnas num√©ricas
def z_score_normalize_dataframe(df):
    # Iterar sobre las columnas del DataFrame
    for column in df.columns:
        if pd.api.types.is_numeric_dtype(df[column]):  # Verificar si la columna es num√©rica
            mean = df[column].mean()
            std_dev = df[column].std()
            df[column] = (df[column] - mean) / std_dev  # Aplicar Z-Score
    return df

# Crear una copia del DataFrame original para normalizarlo
df_normalized = df.copy()

# Aplicar la normalizaci√≥n a todas las columnas num√©ricas
df_normalized = z_score_normalize_dataframe(df_normalized)

# Mostrar los primeros 5 registros del dataset normalizado
print("\nDataset normalizado (Z-Score):\n", df_normalized.head())

# Paso 3: Graficar los histogramas de los atributos antes y despu√©s de la normalizaci√≥n
fig, axes = plt.subplots(2, 4, figsize=(16, 8))  # Cambiamos a 2 filas y 4 columnas (4 atributos)

# Datos originales
df['A'].hist(ax=axes[0, 0], color='blue', alpha=0.6)
axes[0, 0].set_title('Datos Originales - A - c-G.G')

df['B'].hist(ax=axes[0, 1], color='blue', alpha=0.6)
axes[0, 1].set_title('Datos Originales - B - c-G.G')

df['C'].hist(ax=axes[0, 2], color='blue', alpha=0.6)
axes[0, 2].set_title('Datos Originales - C - c-G.G')

df['D'].hist(ax=axes[0, 3], color='blue', alpha=0.6)
axes[0, 3].set_title('Datos Originales - D - c-G.G')

# Datos normalizados
df_normalized['A'].hist(ax=axes[1, 0], color='green', alpha=0.6)
axes[1, 0].set_title('Datos Normalizados - A - c-G.G')

df_normalized['B'].hist(ax=axes[1, 1], color='green', alpha=0.6)
axes[1, 1].set_title('Datos Normalizados - B - c-G.G')

df_normalized['C'].hist(ax=axes[1, 2], color='green', alpha=0.6)
axes[1, 2].set_title('Datos Normalizados - C - c-G.G')

df_normalized['D'].hist(ax=axes[1, 3], color='green', alpha=0.6)
axes[1, 3].set_title('Datos Normalizados - D - c-G.G')

plt.tight_layout()
plt.show()
```
