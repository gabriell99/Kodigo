```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Paso 1: Generar un conjunto de datos con valores aleatorios
np.random.seed(0)
data = np.random.randint(50, 100, size=100)  # Datos enteros entre 50 y 100

# Crear un DataFrame
df = pd.DataFrame({'Original Data': data})

# Paso 2: Función para normalización basada en Z-Score
def z_score_normalization(data):
    mean = np.mean(data)
    std_dev = np.std(data)
    return (data - mean) / std_dev

# Aplicar la normalización Z-Score
df['Z-Score Normalized'] = z_score_normalization(df['Original Data'])

# Paso 3: Graficar los datos originales y los normalizados
plt.figure(figsize=(12, 6))

# Histograma para los datos originales
plt.subplot(1, 2, 1)
plt.hist(df['Original Data'], bins=10, color='blue', alpha=0.7)
plt.title('Datos Originales')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')

# Histograma para los datos normalizados con Z-Score
plt.subplot(1, 2, 2)
plt.hist(df['Z-Score Normalized'], bins=10, color='green', alpha=0.7)
plt.title('Datos transformados (Z-Score)')
plt.xlabel('Valor Z')
plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()

# Paso 4: Verificar la media y desviación estándar de los datos normalizados
mean_normalized = np.mean(df['Z-Score Normalized'])
std_dev_normalized = np.std(df['Z-Score Normalized'])

print(f"Media de los datos transformados: {mean_normalized}")
print(f"Desviación estándar de los datos transformados: {std_dev_normalized}")
```
