```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Paso 1: Generar datos que siguen una distribución normal y no normal
np.random.seed(13) #cambiando 0 por 13 dia cumpleaños
data_normal = np.random.normal(loc=0, scale=1, size=1978)  # media=0, desviación estándar=1... cambiando 1000 * año nacimiento
data_non_normal = np.random.exponential(scale=1, size=1978)  # Distribución exponencial (no normal)... cambiando 1000 * año nacimiento

# Crear un DataFrame para gestionar ambos conjuntos de datos
df = pd.DataFrame({'Normal': data_normal, 'Non-Normal': data_non_normal})

# Paso 3: Gráfico Q-Q (Quantile-Quantile)
plt.figure(figsize=(12, 6))

# Gráfico Q-Q para los datos normales
plt.subplot(1, 2, 1)
stats.probplot(df['Normal'], dist="norm", plot=plt)
plt.title('Gabriel Guzmán - Q-Q Plot - Datos Normales c-G.G')

# Gráfico Q-Q para los datos no normales
plt.subplot(1, 2, 2)
stats.probplot(df['Non-Normal'], dist="norm", plot=plt)
plt.title('Gabriel Guzmán - Q-Q Plot - Datos No Normales c-G.G')

plt.show()

# Paso 4: Prueba de Shapiro-Wilk
shapiro_normal = stats.shapiro(df['Normal'])
shapiro_non_normal = stats.shapiro(df['Non-Normal'])

print(f'c-G.G - Prueba Shapiro-Wilk - Datos Normales: Estadístico={shapiro_normal.statistic}, p-valor={shapiro_normal.pvalue}')
print(f'c-G.G - Prueba Shapiro-Wilk - Datos No Normales: Estadístico={shapiro_non_normal.statistic}, p-valor={shapiro_non_normal.pvalue}')

#si el valor de P es menor a 0.05 no hay evidencia para rechazar la hipotesis nula
#si el valor de P esta arriba de 0.05 siguen la distribución normal
```
