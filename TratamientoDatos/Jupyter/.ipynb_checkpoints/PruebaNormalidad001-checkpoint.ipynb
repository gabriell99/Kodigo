```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Paso 1: Generar datos que siguen una distribución normal y no normal
np.random.seed(0)
data_normal = np.random.normal(loc=0, scale=1, size=1000)  # media=0, desviación estándar=1
data_non_normal = np.random.exponential(scale=1, size=1000)  # Distribución exponencial (no normal)

# Crear un DataFrame para gestionar ambos conjuntos de datos
df = pd.DataFrame({'Normal': data_normal, 'Non-Normal': data_non_normal})

# Paso 3: Gráfico Q-Q (Quantile-Quantile)
plt.figure(figsize=(12, 6))

# Gráfico Q-Q para los datos normales
plt.subplot(1, 2, 1)
stats.probplot(df['Normal'], dist="norm", plot=plt)
plt.title('Q-Q Plot - Datos Normales')

# Gráfico Q-Q para los datos no normales
plt.subplot(1, 2, 2)
stats.probplot(df['Non-Normal'], dist="norm", plot=plt)
plt.title('Q-Q Plot - Datos No Normales')

plt.show()

# Paso 4: Prueba de Shapiro-Wilk
shapiro_normal = stats.shapiro(df['Normal'])
shapiro_non_normal = stats.shapiro(df['Non-Normal'])

print(f'Prueba Shapiro-Wilk - Datos Normales: Estadístico={shapiro_normal.statistic}, p-valor={shapiro_normal.pvalue}')
print(f'Prueba Shapiro-Wilk - Datos No Normales: Estadístico={shapiro_non_normal.statistic}, p-valor={shapiro_non_normal.pvalue}')
```
