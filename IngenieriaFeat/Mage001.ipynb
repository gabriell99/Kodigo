Pasos para instalar Mage y crear un proyecto básico en tu entorno local.

### 1. **Instalar Mage**
   
   python -m venv mage_env
source mage_env/bin/activate  # En Windows, usa mage_env\Scripts\activate
pip install mage
   
   - Asegúrate de que tienes **Python 3.7 o superior** y **pip** instalado.
   - Luego, abre una terminal y ejecuta:
     ```bash
     python -m venv mage_env
     ```
     
     ```bash
     mage_env\Scripts\activate
     ```

     ```bash
     pip install mage-ai
     ```

   Esto instalará Mage y sus dependencias.

### 2. **Crear un Proyecto de Mage**
   - Una vez que Mage esté instalado, puedes iniciar un proyecto ejecutando:

     ```bash
     mage start my_first_project
     ```

   - Esto creará una carpeta llamada `my_first_project`, donde se almacenarán todos los archivos del proyecto, como pipelines, configuraciones y bloques de datos, considerar pip install setuptools si faltan dependencias
 

### 3. **Iniciar Mage en Modo Interactivo**
   - Navega a la carpeta del proyecto:

     ```bash
     cd my_first_project
     ```

   - Ahora inicia el servidor de Mage con:

     ```bash
     mage start
     ```

   - Mage te mostrará una dirección (normalmente `http://localhost:6789`) en la que puedes abrir su interfaz de usuario en el navegador.

### 4. **Crear un Pipeline Básico**
   - En la interfaz de Mage, ve a **Pipelines** y selecciona **Create Pipeline**.
   - Dale un nombre a tu pipeline, como `basic_pipeline`.
   - Escoge el tipo de pipeline (normalmente **Batch pipeline** para ejecutar procesos por lotes) y selecciona **Create**.

### 5. **Agregar un Bloque Básico (Script Python)**
   - En la vista de tu pipeline recién creado, selecciona **Add Block** y elige **Python Block**.
   - Dale un nombre (por ejemplo, `hello_world`) y selecciona **Create Block**.
   - Mage abrirá un editor donde puedes escribir tu primer bloque de código. Agrega algo simple como:

     ```python
     if 'custom' not in globals():
    from mage_ai.data_preparation.decorators import custom
        if 'test' not in globals():
            from mage_ai.data_preparation.decorators import test


        @custom
        def transform_custom(*args, **kwargs):
            """
            args: The output from any upstream parent blocks (if applicable)

            Returns:
                Anything (e.g. data frame, dictionary, array, int, str, etc.)
            """
            # Specify your custom logic here
            d="Hola"

            return d


        @test
        def test_output(output, *args) -> None:
            """
            Template code for testing the output of the block.
            """
            assert output is not None, 'The output is undefined'

             ```

   - Guarda el bloque.

### 6. **Ejecutar el Pipeline**
   - Una vez que hayas guardado el bloque, vuelve a la vista del pipeline y selecciona **Run Pipeline** o haz clic en el botón de ejecución en el bloque para ejecutar solo ese bloque.
   - En la interfaz de Mage, podrás ver la salida en tiempo real. Deberías ver el mensaje `"Hello, Mage!"` en el registro de ejecución.

### 7. **Cerrar el Servidor Mage**
   - Para detener Mage, simplemente vuelve a tu terminal y presiona `Ctrl + C` en la ventana donde lo iniciaste.
