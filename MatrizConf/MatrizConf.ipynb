```python
import random
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
# Establecer una semilla para obtener resultados reproducibles
random.seed(13)
```

```python
# Definir etiquetas reales y predichas usando letras (A = positivo, B = negativo)
labels = ['A', 'B']  # A = positivo, B = negativo

# Generar valores reales y predichos aleatoriamente
y_true = [random.choice(labels) for _ in range(10)]  # Etiquetas reales
y_pred = [random.choice(labels) for _ in range(10)]  # Predicciones del modelo
```

```python
# Crear la matriz de confusi√≥n
cm = confusion_matrix(y_true, y_pred, labels=labels)
```

Una matriz de confusi√≥n es una tabla que se utiliza para evaluar el rendimiento de un modelo de clasificaci√≥n. Proporciona una representaci√≥n detallada de cu√°ntas predicciones fueron correctas o incorrectas, distribuidas en cuatro categor√≠as principales. Se llama "matriz de confusi√≥n" porque ayuda a visualizar d√≥nde el modelo "se confunde" al clasificar correctamente o incorrectamente.

Estructura de la matriz de confusi√≥n:
En una clasificaci√≥n binaria (donde hay dos clases: positivo y negativo), la matriz de confusi√≥n se organiza de la siguiente forma:



Elementos clave:
* True Positive (TP) (Verdaderos Positivos):
Son los casos en los que el modelo predice "positivo" y la realidad tambi√©n es "positivo". Es decir, el modelo acert√≥ al identificar correctamente algo que pertenece a la clase positiva.

* True Negative (TN) (Verdaderos Negativos):
Son los casos en los que el modelo predice "negativo" y la realidad tambi√©n es "negativo". El modelo acert√≥ al identificar correctamente algo que pertenece a la clase negativa.

* False Positive (FP) (Falsos Positivos, tambi√©n llamados "falsos alarmas"):
El modelo predice "positivo", pero en realidad es "negativo". Es decir, el modelo clasific√≥ incorrectamente algo como positivo.

* False Negative (FN) (Falsos Negativos):
El modelo predice "negativo", pero en realidad es "positivo". Es un error donde el modelo no detecta correctamente algo que pertenece a la clase positiva.

```python
# Mostrar los resultados
print("Etiquetas Reales:", y_true)
print("Predicciones del Modelo:", y_pred)

print("\nMatriz de Confusi√≥n:")
print(f"      {labels[0]}   {labels[1]}")
print(f"{labels[0]}   {cm[0][0]}    {cm[0][1]}")
print(f"{labels[1]}   {cm[1][0]}    {cm[1][1]}")

```

C√°lculo del Accuracy:
El accuracy es simplemente el n√∫mero de predicciones correctas sobre el n√∫mero total de predicciones. Se calcula como:





```python
# Calcular el accuracy
accuracy = accuracy_score(y_true, y_pred)
print(f"\nAccuracy: {accuracy * 100:.2f}%")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Crear el heatmap de la matriz de confusi√≥n
plt.figure(figsize=(2,2))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=labels, yticklabels=labels)

# T√≠tulos y etiquetas
plt.title(f'Gabriel Guzm√°n [13] - Matriz de Confusi√≥n - Accuracy: {accuracy * 100:.2f}% - c-G.G')
plt.xlabel('Predicci√≥n')
plt.ylabel('Real')

# Mostrar el gr√°fico
plt.show()
```
